```python
import os
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable
import numpy as np
```

Reading oxygen concentration and computing the average for each time step 

```python
folder_path = 'OutputValuesOxygena'


file_prefix = "Out"
file_suffix = ".txt"

# List to hold the mean of the data from each file
average_oxy = []

# Loop through the file numbers
for i in range(101):  # 0 to 100 inclusive
    file_name = f"{file_prefix}{i}{file_suffix}"
    file_path = os.path.join(folder_path, file_name)
    
    # Check if the file exists to avoid errors
    if os.path.isfile(file_path):
        # Load the data from the file
        data = np.loadtxt(file_path)
        
        # Compute the mean of the data
        oxy_mean = np.mean(data)
        
        # Append the mean to the list
        average_oxy.append(oxy_mean)

```

```python
import numpy as np
import matplotlib.pyplot as plt

file_path = 'l2'  # Change this to your file path

# Read the data from the file and store it in two lists
c = []
lmbd = []

with open(file_path, 'r') as file:
    for line in file:
        columns = line.split()
        c.append(float(columns[0]))
        lmbd.append(float(columns[1]))

# Convert lists to numpy arrays
c = np.array(c)
lmbd = np.array(lmbd)

# Find the index of the first value in the second column that is greater than zero
first_positive_index = np.argmax(lmbd > 0)

# Get the corresponding values from the arrays
if lmbd[first_positive_index] > 0:
    first_positive_lambda = lmbd[first_positive_index]
    corresponding_c = c[first_positive_index]
    print(f"The first index with a value in the second column greater than zero is: {first_positive_index}")
    print(f"The first value in the second column greater than zero is: {first_positive_lambda}")
    print(f"The corresponding value in the first column is: {corresponding_c}")
else:
    print("No positive values found in the second column.")

# Plotting the data
plt.plot(c, lmbd, label='Lambda vs Oxygen')
plt.axvline(x=corresponding_c, color='r', linestyle='--', label=f'First positive lambda at c={corresponding_c}')
plt.grid()
plt.title("Lambda vs Oxygen")
plt.xlabel("Oxygen concentration")
plt.ylabel("Lambda")
plt.legend()
plt.savefig("Test1/lmbd.png")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

file_path = 'l2'  # Change this to your file path

# Read the data from the file and store it in two lists
c = []
lmbd = []

with open(file_path, 'r') as file:
    for line in file:
        columns = line.split()
        c.append(float(columns[0]))
        lmbd.append(float(columns[1]))

# Convert lists to numpy arrays
c = np.array(c)
lmbd = np.array(lmbd)

# Find the index of the first value in the second column that is greater than zero
first_positive_index = np.argmax(lmbd > 0)

# Get the corresponding values from the arrays
if lmbd[first_positive_index] > 0:
    first_positive_lambda = lmbd[first_positive_index]
    corresponding_c = c[first_positive_index]
    print(f"The first index with a value in the second column greater than zero is: {first_positive_index}")
    print(f"The first value in the second column greater than zero is: {first_positive_lambda}")
    print(f"The corresponding value in the first column is: {corresponding_c}")
else:
    print("No positive values found in the second column.")

# Plotting the data with improved styling
plt.figure(figsize=(7, 5))
plt.plot(c, lmbd, label='Lambda vs Oxygen', color='b', linewidth=2)
plt.axvline(x=corresponding_c, color='r', linestyle='--', linewidth=2, label=f'First positive lambda at c={corresponding_c}')
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.title("Dependence of Lambda on Oxygen Concentration", fontsize=13)
plt.xlabel("Oxygen Concentration", fontsize=13)
plt.ylabel("Lambda", fontsize=13)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=12, loc='lower right')
plt.gca().spines['top'].set_linewidth(1)    # Top axis line
plt.gca().spines['right'].set_linewidth(1)  # Right axis line
plt.gca().spines['bottom'].set_linewidth(1) # Bottom axis line
plt.gca().spines['left'].set_linewidth(1)   # Left axis line
plt.tick_params(axis='both', which='major', width=1)
plt.tight_layout()
plt.show()
```
